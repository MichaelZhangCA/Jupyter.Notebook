```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(index=ema12.index)
df['short_ema'] = ema12.ema
df['long_ema'] = ema26.ema
df['macd'] = ema12 - ema26
df['macd_ema'] = xema(df.macd,9)
df['macd_histogram'] = df.macd - df.macd_ema
df.head(30)
```

```python
def xema(x, period):
    
    first_nn = df.index.get_loc(x.first_valid_index())
    
    # roll the first x days for fist SMA start point
    ema_rolling = x[:period + first_nn].rolling(period)

    # combine the first SMA with the rest close adj price
    ema_con = pd.concat([ema_rolling.mean(), x[period + first_nn:]])
    #ema_con.head(30)
    #df['ema_con'] = ema_con
    ema = ema_con.ewm(span=period, adjust=False).mean()
    #return df

    return ema
```
