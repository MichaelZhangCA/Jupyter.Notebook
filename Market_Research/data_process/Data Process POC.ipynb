```python
import pandas as pd
import numpy as np
import datetime

from crossreference import *
import repository.stockpricerepo as pricerepo
from codetimer import CodeTimer
```

```python
with CodeTimer() as ct:
    fullprice = pricerepo.get_stock_adjdata('GOOGL', datetime.date(1900,1,1))

fullprice.head()
```

```python
shortprice = pricerepo.get_stock_adjdata('GOOGL', datetime.date(2010,1,1))
shortprice.head()
```

```python
import indicator.common_algorithm as comalgo
```

```python
period = 20
fullema = comalgo.xema(fullprice.adj_close, period)
shortema = comalgo.xema(shortprice.adj_close, period)
```

```python
df = pd.DataFrame(index = shortema.index)
df['shortema'] = shortema
df['fullema'] = fullema
```

```python
df['delta'] = df.fullema - df.shortema
df['per'] = df.delta / df.fullema 
df['threshold'] = 0.01

df = df.dropna()

print(str(period) + '::' + str(len(df[abs(df.per) >df.threshold])))
df.head(20)
```

```python
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
trace_ema = go.Bar(x=df.index, y=df.per, name='EMA Diff', showlegend=False)
fig = go.Figure(data=[trace_ema])
# standalone chart
plotly.offline.plot(fig)
```
