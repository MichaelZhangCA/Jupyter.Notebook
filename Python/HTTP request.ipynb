```python
import requests
# res = requests.get('https://stackoverflow.com/questions/26000336')
stk = requests.get('https://api.iextrading.com/1.0/stock/aapl/chart/1y')
```

```python
stk.json()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(stk.json())
df
```

```python
# setup chart format
ax = df[["close"]].plot.line(grid='true', figsize=(18,10), title='Close chart', 
                                          marker=".", markersize=10, fontsize=18)
ax.set_xlabel('tenor', fontsize=18)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
from datetime import datetime
```

```python
trace = go.Candlestick(x=df.index, open=df.open, high=df.high, low=df.low, close=df.close)
```

```python
data = [trace]
layout = {
    'title': 'The Great Recession',
    'yaxis': {'title': 'AAPL Stock'},
    'shapes': [{
        'x0': '2016-12-09', 'x1': '2016-12-09',
        'y0': 0, 'y1': 1, 'xref': 'x', 'yref': 'paper',
        'line': {'color': 'rgb(30,30,30)', 'width': 1}
    }],
    'annotations': [{
        'x': '2016-12-09', 'y': 0.05, 'xref': 'x', 'yref': 'paper',
        'showarrow': False, 'xanchor': 'left',
        'text': 'Increase Period Begins'
    }]
}
```

```python
fig = dict(data=data, layout=layout)
py.iplot(fig, filename='appl-recession-candlestick')
```

```python
import plotly
plotly.offline.init_notebook_mode(connected=True) 
plotly.offline.iplot(fig)
```
