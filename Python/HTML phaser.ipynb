```python
import requests

wiki_html = requests.get('http://en.wikipedia.org/wiki/{}'.format('NASDAQ-100'))
```

```python
from bs4 import BeautifulSoup

wiki_soup = BeautifulSoup(wiki_html.content, "html.parser")
symbol_table = wiki_soup.find(attrs={'class': 'div-col columns column-count column-count-2'})

symbol_table
```

```python
for sym in symbol_table.find_all("li"):
    print(sym.contents[1].split('(')[1].replace(")",""))
    print(sym.find('a').text)
```
